```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.boibazar.com/')
response.headers
```

```python
html_data = BeautifulSoup(response.text, 'html.parser')
html_data.title.string
```

```python
print(html_data.prettify())
```

```python
books = html_data.find_all(class_='themeD_product_image_container')
# books.a.img['src']

for i in books:
    print(i.a.img['src'])
```
